A1. Scores

```python
sc = [float(x) for x in input("Input all scores: ").split()];print(f"Number of students = {len(sc)}\nMin score = {min(sc)}\nMax score = {max(sc)}\nAverage = {sum(sc)/len(sc)}")
```

A2. Student Information

```python
i,f,l = input("Enter your code and name: ").split();print(f"Your name is {f} {l}.\nYou are a student year {64-int(i[:2])+1}.")
```

A3. Count distinct letters

```python
s1=set(input("Input a string: "))-{" "};print(f"Count distinct letters = {len(s1)}")
```

A4. Series_A

```python
n=int(input("n = "));lst=[x for x in range(1,1000+1) if x%n==0];print(f"count = {len(lst)}\nlast = {lst[-1]}\n{lst}")
```
